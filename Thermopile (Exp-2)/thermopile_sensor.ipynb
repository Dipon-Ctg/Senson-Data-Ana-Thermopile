# DC offset removal , Spectral analysis & Histogram analysis

```python
import matplotlib.pyplot as plt
from scipy import signal
import pandas as pd
import numpy as np
```

```python
dataFile = pd.read_csv('dataSet_1kHz_5V.csv', index_col=0)

# Original properties
fs = 1000 
dc_offset = 0.35
t = dataFile.index.values
original_signal = dataFile['Voltage']

# DC offset
signal_without_offset = original_signal - original_signal.mean()

# High-pass filter
nyquist = 0.5 * fs
cutoff_freq = 300
order = 1

# Filter
b, a = signal.butter(order, cutoff_freq/nyquist, btype='high', analog=False)
filtered_signal = signal.filtfilt(b, a, signal_without_offset)

plt.subplot(2, 1, 2)
plt.plot(t, filtered_signal, label='1kHz Filtered Signal',color = "gray")
plt.title('DC Offset Removal signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.tight_layout()
plt.show()


# Spectral Analysis
fft_result = np.fft.fft(filtered_signal)
fft_freq = np.fft.fftfreq(len(fft_result), d=1/fs)

plt.figure(figsize=(7, 3))
plt.plot(fft_freq, np.abs(fft_result),color = "gray")
plt.title('Spectral Analysis [fs = 1kHz]')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()



# Histogram Analysis
plt.subplot(3, 1, 3)
plt.hist(filtered_signal, bins='auto', color='gray', alpha=0.7, rwidth=0.85)
plt.title('Histogram of 1kHz')
plt.xlabel('Amplitude')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
